```python
import re
from operator import add
```

```python
# Function to update position and velocity of a particle

def update(particle):
    particle[1] = list(map(add,particle[1],particle[2]))
    particle[0] = list(map(add,particle[0],particle[1]))
    return particle
```

```python
# Read each line in input file. Create
# a list of particle with position, velocity and
# acceleration of each
# Also create lists of absolute distance and velocity

def load_data():
    with open('day20.txt') as f:
        data = f.readlines()

    data = [x.strip() for x in data]




    ps = []

    for i,line in enumerate(data):
        m = re.match(r"p=<\s?(-?\d+),(-?\d+),(-?\d+)>, v=<\s?(-?\d+),(-?\d+),(-?\d+)>, a=<\s?(-?\d+),(-?\d+),(-?\d+)>",line)
        p = [m.group(1),m.group(2),m.group(3)]
        v = [m.group(4),m.group(5),m.group(6)]
        a = [m.group(7),m.group(8),m.group(9)]
        ps.append([p,v,a])

    ps = [ [ [int(z) for z in x] for x in y] for y in ps]

    ds = [sum([abs(y) for y in x[0]]) for x in ps]
    vs = [sum([abs(y) for y in x[1]]) for x in ps]
    
    return ps,ds,vs
```

```python
# Part 1: Update particles iteratively. Keep
# track of the closest particle.
# This is kind of hacky, but after a bunch of iterations
# if the same particle is staying the closest return it

ps,ds,vs = load_data()
closest = []
while True:
    #print('*****')
    ps = [update(x) for x in ps]
    ds = [sum([abs(y) for y in x[0]]) for x in ps]
    vs = [sum([abs(y) for y in x[1]]) for x in ps]
    
    
    closest.append(ds.index(min(ds)))
    
    
    if  len(closest)>4000 and len(set(closest[-100:])) == 1:
        print(closest[-1])
        break

```
